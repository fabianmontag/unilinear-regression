```python
import numpy as np
import matplotlib.pyplot as plt
import random

# data used from https://www.kaggle.com/datasets/yasserh/housing-prices-dataset
# price, area
houseData = np.array([
[13300000.,7420],
[12250000,8960],
[12250000,9960],
[12215000,7500],
[11410000,7420],
[10850000,7500],
[10150000,8580],
[10150000,16200],
[9870000,8100],
[9800000,5750],
[9800000,13200],
[9681000,6000],
[9310000,6550],
[9240000,3500],
[9240000,7800],
[9100000,6000],
[9100000,6600],
[8960000,8500],
[8890000,4600],
[8855000,6420],
[8750000,4320],
[8680000,7155],
[8645000,8050],
[8645000,4560],
[8575000,8800],
[8540000,6540],
[8463000,6000],
[8400000,8875],
[8400000,7950],
[8400000,5500],
[8400000,7475],
[8400000,7000],
[8295000,4880],
[8190000,5960],
[8120000,6840],
[8080940,7000],
[8043000,7482],
[7980000,9000],
[7962500,6000],
[7910000,6000],
[7875000,6550],
[7840000,6360],
[7700000,6480],
[7700000,6000],
[7560000,6000],
[7560000,6000],
[7525000,6000],
[7490000,6600],
[7455000,4300],
[7420000,7440],
[7420000,7440],
[7420000,6325],
[7350000,6000],
[7350000,5150],
[7350000,6000],
[7350000,6000],
[7343000,11440],
[7245000,9000],
[7210000,7680],
[7210000,6000],
[7140000,6000],
[7070000,8880],
[7070000,6240],
[7035000,6360],
[7000000,11175],
[6930000,8880],
[6930000,13200],
[6895000,7700],
[6860000,6000],
[6790000,12090],
[6790000,4000],
[6755000,6000],
[6720000,5020],
[6685000,6600],
[6650000,4040],
[6650000,4260],
[6650000,6420],
[6650000,6500],
[6650000,5700],
[6650000,6000],
[6629000,6000],
[6615000,4000],
[6615000,10500],
[6580000,6000],
[6510000,3760],
[6510000,8250],
[6510000,6670],
[6475000,3960],
[6475000,7410],
[6440000,8580],
[6440000,5000],
[6419000,6750],
[6405000,4800],
[6300000,7200],
[6300000,6000],
[6300000,4100],
[6300000,9000],
[6300000,6400],
[6293000,6600],
[6265000,6000],
[6230000,6600],
[6230000,5500],
[6195000,5500],
[6195000,6350],
[6195000,5500],
[6160000,4500],
[6160000,5450],
[6125000,6420],
[6107500,3240],
[6090000,6615],
[6090000,6600],
[6090000,8372],
[6083000,4300],
[6083000,9620],
[6020000,6800],
[6020000,8000],
[6020000,6900],
[5950000,3700],
[5950000,6420],
[5950000,7020],
[5950000,6540],
[5950000,7231],
[5950000,6254],
[5950000,7320],
[5950000,6525],
[5943000,15600],
[5880000,7160],
[5880000,6500],
[5873000,5500],
[5873000,11460],
[5866000,4800],
[5810000,5828],
[5810000,5200],
[5810000,4800],
[5803000,7000],
[5775000,6000],
[5740000,5400],
[5740000,4640],
[5740000,5000],
[5740000,6360],
[5740000,5800],
[5652500,6660],
[5600000,10500],
[5600000,4800],
[5600000,4700],
[5600000,5000],
[5600000,10500],
[5600000,5500],
[5600000,6360],
[5600000,6600],
[5600000,5136],
[5565000,4400],
[5565000,5400],
[5530000,3300],
[5530000,3650],
[5530000,6100],
[5523000,6900],
[5495000,2817],
[5495000,7980],
[5460000,3150],
[5460000,6210],
[5460000,6100],
[5460000,6600],
[5425000,6825],
[5390000,6710],
[5383000,6450],
[5320000,7800],
[5285000,4600],
[5250000,4260],
[5250000,6540],
[5250000,5500],
[5250000,10269],
[5250000,8400],
[5250000,5300],
[5250000,3800],
[5250000,9800],
[5250000,8520],
[5243000,6050],
[5229000,7085],
[5215000,3180],
[5215000,4500],
[5215000,7200],
[5145000,3410],
[5145000,7980],
[5110000,3000],
[5110000,3000],
[5110000,11410],
[5110000,6100],
[5075000,5720],
[5040000,3540],
[5040000,7600],
[5040000,10700],
[5040000,6600],
[5033000,4800],
[5005000,8150],
[4970000,4410],
[4970000,7686],
[4956000,2800],
[4935000,5948],
[4907000,4200],
[4900000,4520],
[4900000,4095],
[4900000,4120],
[4900000,5400],
[4900000,4770],
[4900000,6300],
[4900000,5800],
[4900000,3000],
[4900000,2970],
[4900000,6720],
[4900000,4646],
[4900000,12900],
[4893000,3420],
[4893000,4995],
[4865000,4350],
[4830000,4160],
[4830000,6040],
[4830000,6862],
[4830000,4815],
[4795000,7000],
[4795000,8100],
[4767000,3420],
[4760000,9166],
[4760000,6321],
[4760000,10240],
[4753000,6440],
[4690000,5170],
[4690000,6000],
[4690000,3630],
[4690000,9667],
[4690000,5400],
[4690000,4320],
[4655000,3745],
[4620000,4160],
[4620000,3880],
[4620000,5680],
[4620000,2870],
[4620000,5010],
[4613000,4510],
[4585000,4000],
[4585000,3840],
[4550000,3760],
[4550000,3640],
[4550000,2550],
[4550000,5320],
[4550000,5360],
[4550000,3520],
[4550000,8400],
[4543000,4100],
[4543000,4990],
[4515000,3510],
[4515000,3450],
[4515000,9860],
[4515000,3520],
[4480000,4510],
[4480000,5885],
[4480000,4000],
[4480000,8250],
[4480000,4040],
[4473000,6360],
[4473000,3162],
[4473000,3510],
[4445000,3750],
[4410000,3968],
[4410000,4900],
[4403000,2880],
[4403000,4880],
[4403000,4920],
[4382000,4950],
[4375000,3900],
[4340000,4500],
[4340000,1905],
[4340000,4075],
[4340000,3500],
[4340000,6450],
[4319000,4032],
[4305000,4400],
[4305000,10360],
[4277000,3400],
[4270000,6360],
[4270000,6360],
[4270000,4500],
[4270000,2175],
[4270000,4360],
[4270000,7770],
[4235000,6650],
[4235000,2787],
[4200000,5500],
[4200000,5040],
[4200000,5850],
[4200000,2610],
[4200000,2953],
[4200000,2747],
[4200000,4410],
[4200000,4000],
[4200000,2325],
[4200000,4600],
[4200000,3640],
[4200000,5800],
[4200000,7000],
[4200000,4079],
[4200000,3520],
[4200000,2145],
[4200000,4500],
[4193000,8250],
[4193000,3450],
[4165000,4840],
[4165000,4080],
[4165000,4046],
[4130000,4632],
[4130000,5985],
[4123000,6060],
[4098500,3600],
[4095000,3680],
[4095000,4040],
[4095000,5600],
[4060000,5900],
[4060000,4992],
[4060000,4340],
[4060000,3000],
[4060000,4320],
[4025000,3630],
[4025000,3460],
[4025000,5400],
[4007500,4500],
[4007500,3460],
[3990000,4100],
[3990000,6480],
[3990000,4500],
[3990000,3960],
[3990000,4050],
[3920000,7260],
[3920000,5500],
[3920000,3000],
[3920000,3290],
[3920000,3816],
[3920000,8080],
[3920000,2145],
[3885000,3780],
[3885000,3180],
[3850000,5300],
[3850000,3180],
[3850000,7152],
[3850000,4080],
[3850000,3850],
[3850000,2015],
[3850000,2176],
[3836000,3350],
[3815000,3150],
[3780000,4820],
[3780000,3420],
[3780000,3600],
[3780000,5830],
[3780000,2856],
[3780000,8400],
[3773000,8250],
[3773000,2520],
[3773000,6930],
[3745000,3480],
[3710000,3600],
[3710000,4040],
[3710000,6020],
[3710000,4050],
[3710000,3584],
[3703000,3120],
[3703000,5450],
[3675000,3630],
[3675000,3630],
[3675000,5640],
[3675000,3600],
[3640000,4280],
[3640000,3570],
[3640000,3180],
[3640000,3000],
[3640000,3520],
[3640000,5960],
[3640000,4130],
[3640000,2850],
[3640000,2275],
[3633000,3520],
[3605000,4500],
[3605000,4000],
[3570000,3150],
[3570000,4500],
[3570000,4500],
[3570000,3640],
[3535000,3850],
[3500000,4240],
[3500000,3650],
[3500000,4600],
[3500000,2135],
[3500000,3036],
[3500000,3990],
[3500000,7424],
[3500000,3480],
[3500000,3600],
[3500000,3640],
[3500000,5900],
[3500000,3120],
[3500000,7350],
[3500000,3512],
[3500000,9500],
[3500000,5880],
[3500000,12944],
[3493000,4900],
[3465000,3060],
[3465000,5320],
[3465000,2145],
[3430000,4000],
[3430000,3185],
[3430000,3850],
[3430000,2145],
[3430000,2610],
[3430000,1950],
[3423000,4040],
[3395000,4785],
[3395000,3450],
[3395000,3640],
[3360000,3500],
[3360000,4960],
[3360000,4120],
[3360000,4750],
[3360000,3720],
[3360000,3750],
[3360000,3100],
[3360000,3185],
[3353000,2700],
[3332000,2145],
[3325000,4040],
[3325000,4775],
[3290000,2500],
[3290000,3180],
[3290000,6060],
[3290000,3480],
[3290000,3792],
[3290000,4040],
[3290000,2145],
[3290000,5880],
[3255000,4500],
[3255000,3930],
[3234000,3640],
[3220000,4370],
[3220000,2684],
[3220000,4320],
[3220000,3120],
[3150000,3450],
[3150000,3986],
[3150000,3500],
[3150000,4095],
[3150000,1650],
[3150000,3450],
[3150000,6750],
[3150000,9000],
[3150000,3069],
[3143000,4500],
[3129000,5495],
[3118850,2398],
[3115000,3000],
[3115000,3850],
[3115000,3500],
[3087000,8100],
[3080000,4960],
[3080000,2160],
[3080000,3090],
[3080000,4500],
[3045000,3800],
[3010000,3090],
[3010000,3240],
[3010000,2835],
[3010000,4600],
[3010000,5076],
[3010000,3750],
[3010000,3630],
[3003000,8050],
[2975000,4352],
[2961000,3000],
[2940000,5850],
[2940000,4960],
[2940000,3600],
[2940000,3660],
[2940000,3480],
[2940000,2700],
[2940000,3150],
[2940000,6615],
[2870000,3040],
[2870000,3630],
[2870000,6000],
[2870000,5400],
[2852500,5200],
[2835000,3300],
[2835000,4350],
[2835000,2640],
[2800000,2650],
[2800000,3960],
[2730000,6800],
[2730000,4000],
[2695000,4000],
[2660000,3934],
[2660000,2000],
[2660000,3630],
[2660000,2800],
[2660000,2430],
[2660000,3480],
[2660000,4000],
[2653000,3185],
[2653000,4000],
[2604000,2910],
[2590000,3600],
[2590000,4400],
[2590000,3600],
[2520000,2880],
[2520000,3180],
[2520000,3000],
[2485000,4400],
[2485000,3000],
[2450000,3210],
[2450000,3240],
[2450000,3000],
[2450000,3500],
[2450000,4840],
[2450000,7700],
[2408000,3635],
[2380000,2475],
[2380000,2787],
[2380000,3264],
[2345000,3640],
[2310000,3180],
[2275000,1836],
[2275000,3970],
[2275000,3970],
[2240000,1950],
[2233000,5300],
[2135000,3000],
[2100000,2400],
[2100000,3000],
[2100000,3360],
[1960000,3420],
[1890000,1700],
[1890000,3649],
[1855000,2990],
[1820000,3000],
[1767150,2400],
[1750000,3620],
[1750000,2910],
[1750000,3850],
])


random.shuffle(houseData)


# predict SqFt with house price

max = houseData[0][0]
min = houseData[0][0]
# feature normilazation of price
for i in range(1, houseData.shape[0]):
    if (max < houseData[i][0]): max = houseData[i][0]
    if (min > houseData[i][0]): min = houseData[i][0]

for i in range(houseData.shape[0]):
    houseData[i][0] = (houseData[i][0]-min)/(max-min)

# indices for feature and label 
featureIndex = 0; # house price
labelIndex = 1; # area

totalM = houseData.shape[0]
testM = 70
trainM = totalM-testM

test_x = np.zeros(testM)
test_y = np.zeros(testM)

for i in range(testM):
    test_x[i] = houseData[trainM+i][featureIndex]
    test_y[i] = houseData[trainM+i][labelIndex]

train_x = np.zeros(trainM)
train_y = np.zeros(trainM)

for i in range(trainM):
    train_x[i] = houseData[i][featureIndex]
    train_y[i] = houseData[i][labelIndex]

print("house prices")
print(train_x)

print("house SqFt's")
print(train_y)

# scatter training data in plot
plt.title("data + model")
plt.scatter(train_x, train_y)

# model paremters weight w and bias b
w = 0
b = 0

# runs the model function with given parameters weight w and bias b
def runFunction(w, x, b):
    return w * x + b

# calculates the mean square error
def calculateMSE(featuresX, labelsY, m, w, b):
    totalLoss = 0
    for i in range(m):
        totalLoss+= (runFunction(w, featuresX[i], b) - labelsY[i]) ** 2
    return 1/m * totalLoss

print(calculateMSE(train_x, train_y, trainM, w, b))

# math equations
# w = w - alpha * d/dw * 1/2m * (m sigma i:= 1) (fwb(x_i) - y_i)**2
# w = w - alpha * 1 / m * (m sigma i:= 1) (fwb(x_i) - y_i) * x_i
# d*L(w, b)/dw = 1 / m * (m sigma i:= 1) (fwb(x_i) - y_i) * x_i

# b = b - alpha * d/db * 1/2m * (m sigma i:= 1) (fwb(x_i) - y_i)**2
# b = b - alpha * 1 / m * (m sigma i:= 1) (fwb(x_i) - y_i)
# d*L(w, b)/db = 1 / m * (m sigma i:= 1) (fwb(x_i) - y_i)

# calculates the slope
def calculateGradient(featuresX, labelsY, m, w, b):
    dw = 0
    db = 0

    for i in range(m):
        dw += ((w * featuresX[i] + b) - labelsY[i]) * featuresX[i]
        db += ((w * featuresX[i] + b) - labelsY[i])
    
    dw = dw / m
    db = db / m

    return dw, db

w_slope = []
b_slope = []
iters = []
    
# gradient descent function
def gradientDescent(featuresX, labelsY, m, w, b, alpha, iterationCount):
    new_w = 0.2
    new_b = 0.2

    for i in range(iterationCount):
        dw, db = calculateGradient(featuresX, labelsY, m, new_w, new_b)
        
        new_w = new_w - alpha * dw
        new_b = new_b - alpha * db

        iters.append(i)
        w_slope.append(dw)
        b_slope.append(db)

    return [new_w, new_b]

iterations = 300
alpha = 0.1 # "learning rate"

newParameters = gradientDescent(train_x, train_y, trainM, w, b, alpha, iterations)
print(newParameters)

plt.plot([0, 1], [newParameters[0] * 0 + newParameters[1], newParameters[0] * 1 + newParameters[1]])
plt.show()

## testdata
predictions = []
for i in range(testM):
    y_hat = newParameters[0] * test_x[i] + newParameters[1]
    predictions.append(y_hat)
    print(f"price: {test_x[i]*(max-min)+min}, prediction: {y_hat}, target: {test_y[i]}")

plt.title("predictions (red) vs test value (green)")
plt.plot([0, 1], [newParameters[0] * 0 + newParameters[1], newParameters[0] * 1 + newParameters[1]])
plt.scatter(test_x, predictions, marker="x", c="r")
plt.scatter(test_x, test_y, marker="x", c="g")
plt.show()

plt.title("learning curve, weight (orange), bias (blue)")
plt.plot(iters, w_slope, c="orange")
plt.plot(iters, b_slope, c="dodgerblue")
plt.xlabel("Iterations")
plt.ylabel("Slope")
plt.show()
```
